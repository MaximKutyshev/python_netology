### Задание №1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

* если источник traffic_source равен yandex или google, то в source_type ставится organic
* для источников paid и email из России - ставим ad
* для источников paid и email не из России - ставим other
* все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/maksim/Desktop/supreme-memory/доп_материалы_к_пандас_1_и_2/visit_log.csv', sep = ';')
```

```python
df
```

```python
df['traffic_source'].unique()
```

```python
def source_types(row):
    if row['traffic_source'] == 'yandex' or row['traffic_source'] == 'google':
        return 'organic'
    elif row['traffic_source'] == 'paid' or row['traffic_source'] == 'email':
        if row['region'] == 'Russia':
            return 'ad'
        elif row['region'] != 'Russia':
            return 'other'
    else:
        return row['traffic_source']
```

```python
df['source_type'] = df.apply(source_types, axis = 1)
```

```python
df
```

```python
df['source_type'].unique()
```

### Задание №3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.


```python
df_ratings = pd.read_csv('/Users/maksim/Downloads/ml-latest-small/ratings.csv')
df_ratings 
```

```python
data = df_ratings.groupby(['userId']) 

data_count = data['movieId'].count().reset_index()
data_count.rename(columns = {'movieId' : 'count'}, inplace = True)
data_active_users = data_count[data_count['count'] > 100]

data_active_users 
```

```python
data_time_max = data['timestamp'].max().reset_index()
data_time_max.rename(columns = {'timestamp' : 'time_max'}, inplace = True)
```

```python
data_time_max 
```

```python
data_time_min = data['timestamp'].min().reset_index()
data_time_min.rename(columns = {'timestamp' : 'time_min'}, inplace = True)
```

```python
data_time_min
```

```python
data_time = data_time_max.merge(data_time_min, how = 'left', on = 'userId')
```

```python
data_time
```

```python
data_active_users_lifecycle = data_active_users.merge(data_time, how = 'left', on = 'userId')
data_active_users_lifecycle
```

```python
def lifecycle(data):
    return (data['time_max'] - data['time_min'])
```

```python
data_active_users_lifecycle['lifecycle'] = data_active_users_lifecycle.apply(lifecycle, axis = 1)
data_active_users_lifecycle
```

```python
data_active_users_lifecycle['lifecycle'].mean()
```

### Задание №4

Дана статистика услуг перевозок клиентов компании по типам (см. файл “Python_13_join.ipynb” в разделе Материалы для лекции «Продвинутый pandas» ---- Ноутбуки к лекции «Продвинутый pandas»).
Необходимо сформировать две таблицы:
1. таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
2. аналогичную таблицу по типам выручки с указанием адреса клиента

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
client_base_rzd = client_base.merge(rzd, how = 'outer', on = 'client_id')
```

```python
client_base_rzd_auto = client_base_rzd.merge(auto, how = 'outer', on = 'client_id')
```

```python
client_base_all = client_base_rzd_auto.merge(air, how = 'outer', on = 'client_id')
```

```python
client_base_all
```

```python
client_base_all.pivot_table(index = 'client_id')
```

```python
client_base_all.pivot_table(index = ['client_id', 'address'])
```
