# Визуализация данных

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 13,8
```

```python
df_2016 = pd.read_csv('/Users/maksim/Desktop/archive/2016.csv')
df = df_2016.rename(columns = {'Trust (Government Corruption)': 'Trust'})
df.head(10)
```

# Bar Chart

```python
df.groupby('Region')['Happiness Score'].mean().plot(kind = 'bar',
                                                        title = 'Средний уровень счастья для каждого региона',
                                                        xlabel = 'Region',
                                                        ylabel = 'Average Happiness Score')
```

Исходя из построенной визуализации видно, что наибольший уровень счастья соответсвует странам Северной Америки, Западной Европы и Австралии с Новой Зеландией.

# Точечная диаграмма 

```python
df.plot(kind = 'scatter',
       x = 'Trust',
       y = 'Freedom',
       title = 'Зависимость между оценкой свободы и доверием к властям')
```

Данный график показывает, что в целом уровень доверия властям не влияет на уровень свободы в стране.

```python
df.plot(kind = 'scatter',
       x = 'Happiness Score',
       y = 'Health (Life Expectancy)',
       title = 'Зависимость между уровнем счастья и ожидаемой продолжительностью жизни')
```

 На основе построенного точечного графика, можно проследить линейную зависимость между уровнем счастья в стране и ожидаемой продолжительности жизни. Для того чтобы увеличить продолжительность жизни нужно уделять внимание не только медицине, но и тем сферам жизни, которые оказывают влияние на уровень счастья в стране.

# PairPlot

```python
sns.pairplot(df[['Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', '']])
```

# heatmap

```python
data_ = df.pivot_table(index = 'Country',
                       columns = 'Region',
                       values = 'Happiness Score',
                       aggfunc = sum ).fillna(0)
```

```python
sns.heatmap(data_)
```

Потренировался в визуализации heatmap. Пробовал различные варианты

# Box Plot

```python
sns.boxplot(x="Region", y="Happiness Score", 
            data = df)

rcParams['figure.figsize'] = 50, 10
```

Данная визуализация дает полное представление об уровне счастья для каждого региона. С помощью boxplot мы видим медианы, размах данного показателя.
